***BASIC CALCULATOR***

This task is to create a basic calculator that takes input from user and do simple arithmetic calculations

```python
print("First enter any two numbers you want to calculate")
print("Then select your mode:'add','sub','mul','div','mod','pow'")
a=float(input("Enter your 1st number: "))
b=float(input("Enter your 2nd number: "))
mode=input("Enter any one mode among the above list: ")
if ( mode=='add'):
  print("The addition of two numbers =",a+b)
elif ( mode=='sub'):
  print("The addition of two numbers =",a-b)
elif ( mode=='mul'):
  print("The multiplication of two numbers =",a*b)
elif ( mode=='div'):
  print("The division of two numbers =",a/b)
elif ( mode=='mod'):
  print("The modulus of two numbers =",a%b)
elif ( mode=='pow'):
  print("The power of two numbers =",pow(a,b))
else:
  print("Please follow the instructions to get desired output")
```
