# ***ROCK PAPER SCISSORS***

***This task is just a demonstration of the rock paper scissors game***

```python
print("WELCOME TO ROCK PAPER SCISSORS  GAME. ENJOY PLAYING!!")
A=input("Want to play ?Type 'YES' or 'NO' :")
a=A.lower()
if(a=="yes"):

  import random

  ug=0
  cg=0

  for i in range(0,5):
    user=input("Select 'rock','paper' or 'scissors':")
    comp=random.choice(['rock','paper','scissors'])
    if(user==comp):
      print("YOU",user)
      print("COMP",comp)
      print("TIE")
    elif(user=='rock' and comp=='scissors'):
      print("YOU",user)
      print("COMP",comp)
      print("YOU","WIN")
      ug+=1
    elif(comp=='rock' and user=='scissors'):
      print("YOU",user)
      print("COMP",comp)
      print("YOU","LOSE")
      cg+=1
    elif(user=='scissors' and comp=='paper'):
      print("YOU",user)
      print("COMP",comp)
      print("YOU","WIN")
      ug+=1
    elif(user=='paper' and comp=='scissors'):
      print("YOU",user)
      print("COMP",comp)
      print("YOU","LOSE")
      cg+=1
    elif(user=='paper' and comp=='rock'):

      print("YOU",user)
      print("COMP",comp)
      print("YOU","WIN")
      ug+=1
    elif(user=='rock' and comp=='paper'):
      print("YOU",user)
      print("COMP",comp)
      print("YOU","LOSE")
      cg+=1
    else:
      print("Please type the given inputs for proper answer")


  if(ug>cg):
    print ("USER POINTS",ug)
    print ("COMP POINTS",cg)
    print("YOU WIN THE MATCH")
  elif(cg>ug):
    print ("USER POINTS",ug)
    print ("COMP POINTS",cg)
    print("COMP WIN THE MATCH")
  else:
    print ("USER POINTS",ug)
    print ("COMP POINTS",cg)
    print("IT'S A DRAW")
else:
  print("Bye!! See you again")
```
